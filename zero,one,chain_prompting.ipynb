```python
!pip install transformers
from transformers import pipeline

# Use a different model that is publicly available on the Hugging Face Model Hub
generator = pipeline("text-generation", model="EleutherAI/gpt-neo-2.7B")

zero_shot_prompt = "what are the benefits of regular exercise?"
result = generator(zero_shot_prompt, max_length=50, do_sample=True, temperature=0.9)
print(result[0]['generated_text'])
```

```python
one_shot_prompt=(
    "Translate English to French: \n"
    "sea otter => loutre de mer\n"
    "Goodbye =>"
    "Hello =>"
    "My name is =>"
)

response=generator(one_shot_prompt,max_length=50)
print(response[0]['generated_text'])
```

```python
chain_of_thought_prompt=("Lets solve this math problem step by step:\n"
                          "what is 15 times 3?\n"
                          "First, i will break it down:\n"
                          "15*3=")
response=generator(chain_of_thought_prompt,max_length=50)
print(response[0]['generated_text'])
```
